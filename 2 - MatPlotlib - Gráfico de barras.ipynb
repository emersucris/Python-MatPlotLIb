```python
# Importando as Bibliotecas do Python.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando os Rótulos e Elementos de Visualização (Dimensão - Label)
labels = ['A1','A2','A3','A4','A5']
```

```python
valores = [10,45,32,56,23]
```

```python
# Plotando o gráfico de barras simples - sem nenhum detalhe visual
plt.bar(labels,valores)
```

```python
# Colocando os valores em cima das barras em forma horizontal
plt.bar(labels,valores,color='red')
plt.xlabel('Código Produto')
plt.ylabel('Valores')
```
