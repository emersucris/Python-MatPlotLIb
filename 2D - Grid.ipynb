```python
# Importando as Bibliotecas.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando a variável de X.
X = np.linspace(0,5,11)
```

```python
# Criando funções de Y(X) e Z(X).
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
# Criando o Grid.
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(b=True , linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

```python
# Alterando a cor de fundo.
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
