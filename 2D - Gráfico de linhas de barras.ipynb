```python
# Importando as Bibliotecas do Python.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando a variável de X
X = np.linspace(0,5,11)
```

```python
# Criando funções de Y(X) e Z(X)
Y = np.power(X-2,2.)
Z = np.exp(X/2.0)
```

```python
# Gráfico de Y em função de X.
plt.plot(X,Y)
plt.show()
```

```python
# Inserindo nomes aos eixos.
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Inserindo titúlo no gráfico.
plt.plot(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
# Dois graficos na mesma imagem.
plt.plot(X,Y)
plt.plot(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Alterando as cores.
plt.plot(X,Y,color='magenta')
plt.plot(X,Z,color='purple')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Inserindo legenda.
plt.plot(X,Y,color='Magenta',label='Y(X)')
plt.plot(X,Z,color='blue',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
