```python
# Importando Bibliotecas.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável de X.
X = np.linspace(0,5,11)
```

```python
# Criando funções de Y(X) e Z(X).
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
# Gráfico de Y em função de X.
plt.scatter(X,Y)
plt.show()
```

```python
# Inserindo nome aos eixos.
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Inserindo título no gráfico.
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
# Dois gráficos na mesma imagem.
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Alterando as cores.
plt.scatter(X,Y,color='cyan')
plt.scatter(X,Z,color='magenta')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Inserindo legenda.
plt.scatter(X,Y,color='cyan',label='Y(X)')
plt.scatter(X,Z,color='magenta',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

```python
# Alterando símbolo.
plt.scatter(X,Y,marker='.')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Dois gráficos co símbolos diferentes.
plt.scatter(X,Y,marker='p')
plt.scatter(X,Z,marker='v')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Alterando o tamanho do ponto.
plt.scatter(X,Y,s=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show() 
```

```python
plt.scatter(X,Y,s=10,label='Y(X)')
plt.scatter(X,Z,s=40,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,s=10,label='Y(X)',alpha=0.8)
plt.scatter(X,Z,s=40,label='Z(X)',alpha=0.4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
